```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 5
theta = np.linspace(0, 2 * np.pi, 500)
theta_point = np.pi / 3
```

```python
x_cardioid = a * (2 * np.cos(theta) - np.cos(2 * theta))
y_cardioid = a * (2 * np.sin(theta) - np.sin(2 * theta))

x_base = a * np.cos(theta)
y_base = a * np.sin(theta)

x_roll_center = 2 * a * np.cos(theta_point)
y_roll_center = 2 * a * np.sin(theta_point)
x_roll = x_roll_center + a * np.cos(theta)
y_roll = y_roll_center + a * np.sin(theta)

px = a * (2 * np.cos(theta_point) - np.cos(2 * theta_point))
py = a * (2 * np.sin(theta_point) - np.sin(2 * theta_point))

plt.figure(figsize=(8, 8))
plt.plot(x_cardioid, y_cardioid, color='red', linewidth=2, label='Кардіоїда')
plt.plot(x_base, y_base, color='blue', linestyle='--', label='Нерухоме коло')
plt.plot(x_roll, y_roll, color='green', linestyle=':', label='Рухоме коло')
plt.plot([x_roll_center, px], [y_roll_center, py], color='black', marker='o')

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Кардіоїда')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.savefig("task4_cardioid.png")
plt.show()
```
