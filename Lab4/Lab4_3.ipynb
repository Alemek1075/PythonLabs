```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lab2_var099.csv')
bins_count = 15
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['n'], bins=bins_count, edgecolor='black', color='skyblue')
plt.xlabel('Розмір даних (n)')
plt.ylabel('Частота (кількість точок)')
plt.title('Розподіл значень аргументу по інтервалах')
plt.grid(axis='y', alpha=0.75)
plt.savefig("task3_histogram.png")
plt.show()
```
