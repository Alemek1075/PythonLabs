```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('lab2_var099.csv')
df = df.sort_values(by='n')
x_data = df['n'].values
y_data = df['t'].values
```

```python
fit_func = lambda n, a, b: a * n * np.log(n + 1) + b
popt, _ = curve_fit(fit_func, x_data, y_data)
y_fit = fit_func(x_data, *popt)

plt.figure(figsize=(10, 6))
plt.plot(x_data, y_data, marker='o', linestyle='-', color='blue', label='Емпіричні дані')
plt.plot(x_data, y_fit, linestyle='--', color='red', label='Апроксимація O(n log n)')
plt.xlabel('Розмір вхідних даних (n)')
plt.ylabel('Час виконання (t)')
plt.title('Оцінка часової складності')
plt.legend()
plt.grid(True)
plt.savefig("task2_complexity.png")
plt.show()
```
