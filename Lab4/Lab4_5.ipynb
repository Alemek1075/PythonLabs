**Умова:** Намалювати тривимірну поверхню (wireframe plot). На зображенні має бути підпис та легенда з рівнянням поверхні. Використання циклів заборонено.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 3.0, 3.0, 3.0
grid_points = 30
u = np.linspace(0, 2 * np.pi, grid_points)
v = np.linspace(1.0, 3.0, grid_points)
U, V = np.meshgrid(u, v)
```

```python
X = a * np.sinh(V) * np.cos(U)
Y = b * np.sinh(V) * np.sin(U)
Z1 = c * np.cosh(V)
Z2 = -c * np.cosh(V)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z1, color='blue', label=r'$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1$')
ax.plot_wireframe(X, Y, Z2, color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Двопорожнинний гіперболоїд')
ax.legend()
plt.savefig("task5_hyperboloid.png")
plt.show()
```
