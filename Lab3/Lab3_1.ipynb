```python
from pathlib import Path
import io
import json
import pandas as pd
import sys
import requests
import bs4
from bs4 import BeautifulSoup as BSoup, SoupStrainer, BeautifulSoup
import html5lib
```

```python
url = "https://www.numbeo.com/cost-of-living/prices_by_country.jsp?displayCurrency=EUR&itemId=12"
r = requests.get(url)
r.status_code
```

```python
fname = Path('data.html')
cheese_data = r.text
with open(fname, "w", encoding="utf-8") as f:
    f.write(cheese_data)
```

```python
with open(fname,"r" ,encoding="utf-8") as f:
    data = f.read()
data
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, "html.parser")
```

```python
tables = soup.find_all("table")
table_index = None
for index,table in enumerate(tables):
    if'Local Cheese ' in str(table):
        table_index = index
        break
table_index
```

```python
cheese = pd.read_html(io.StringIO(str(tables[table_index])), flavor = 'bs4')[0]
cheese
```

```python
min_val = cheese["Local Cheese (1 kg)"].min()
min_row = cheese[cheese["Local Cheese (1 kg)"] == min_val]
min_row
```

```python
max_val = cheese["Local Cheese (1 kg)"].max()
max_row = cheese[cheese["Local Cheese (1 kg)"] == max_val]
max_row
```
