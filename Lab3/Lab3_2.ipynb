```python
import requests
from bs4 import BeautifulSoup

url = "https://en.cppreference.com/w/cpp/container/deque"
soup = BeautifulSoup(requests.get(url).text, "html.parser")
```

```python
t_table = soup.find(id="Member_types").find_next("table")
print("Member types:\n" + ", ".join([c.text for c in t_table.find_all("code")]))
```

```python
print("Member functions:")
f_table = soup.find(id="Member_functions").find_next("table")
for tr in f_table.find_all("tr"):
    if tr.h5:
        print(f"\n{tr.h5.text}:", end=" ")
    else:
        funcs = [s.text for s in tr.select(".t-lines > span") if s.text]
        if funcs:
            print(", ".join(funcs), end=", ")
```

```python
n_table = soup.find(id="Non-member_functions").find_next("table")
n_funcs = set(s.text.replace("(std::deque)", "") for s in n_table.select(".t-lines > span") if s.text)
print("\nNon-member functions:\n" + ", ".join(n_funcs))
```

```python
c_soup = BeautifulSoup(requests.get(url + "/deque").text, "html.parser")
ctors = c_soup.select(".t-dcl-begin tr.t-dcl")

print(f"Конструкторів: {len(ctors)}")

for tr in ctors:
    num = tr.find("td").text.strip(" ()")
    mark = tr.select_one(".t-mark-rev")
    std = mark.text.strip(" ()").replace("since ", "") if mark else "C++98"
    print(f"{num}: {std}")

cxx23 = [tr.find("td").text.strip(" ()") for tr in ctors]
print(f"C++23: {', '.join(cxx23)}")
```
