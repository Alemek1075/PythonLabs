# pdf2image

[pdf2image](https://pypi.org/project/pdf2image/#description) &mdash; бібліотека для перетворення pdf-файлів у зображення.

Встановлення: 
<br>
<code>pip install pdf2image</code>

Використання:
<br>
<code>import pdf2image</code>


```python
from pathlib import Path
import pdf2image as m
from PIL import Image
import shutil
```

```python
root = Path('.')
data = root.joinpath('data')
pdfsample = data.joinpath('104.pdf')
```

[convert_from_path](https://pypi.org/project/pdf2image/#description) &mdash; перетворює pdf-файл на колекцію зображень сторінок. Зображення знаходяться у внутрішньому форматі бібліотеки <code>pillow</code>.

```python
tmp = data.joinpath('tmp')
pages_dir = data.joinpath('pages_02')
tmp.mkdir(exist_ok=True)
pages_dir.mkdir(exist_ok=True)
pages = m.convert_from_path(pdfsample, output_folder=tmp)
for i, im in enumerate(pages):
    im.save(pages_dir.joinpath(f'{i+1:02}.jpg'))
print(f'{i+1} pages')
shutil.rmtree(tmp)
```
