```python
from docx import Document
import docx
from pathlib import Path
```

```python
root = Path('.')
sample = root.joinpath('data', 'bac_inf_opp_2025_2026_для друку_12_05.docx')
```

```python
doc = Document(sample)
```

```python
i =0
for el in doc.iter_inner_content():
    print(el)
    i+=1
    if i>35 and el.text: break
    # break
```

```python
el.text, i
```

```python
for el in doc.tables:
    print(el)
```

```python
for row in doc.tables[-2].rows:
    print(row)
    for cell in row.cells:
        print(cell.grid_span, cell.text, end= ' ')
    print()
```
