```python
import openpyxl
from openpyxl import load_workbook, Workbook
from openpyxl.styles import PatternFill
from openpyxl.utils import get_column_letter
from openpyxl.styles import Alignment

from pathlib import Path
print(f'openpyxl {openpyxl.__version__}')
```

```python
fillGrey = PatternFill(start_color="d3d3d3", fill_type = "solid")
fillBlue = PatternFill(start_color="ADD8E6", fill_type = "solid")
CENTER = Alignment(horizontal='center')
LEFT = Alignment(horizontal='left')
```

```python
root = Path('.')
data = root.joinpath('data', '2024 Інформатика.xlsx')
sample = root.joinpath('data', 'sample.xlsx')
```

```python
wb = Workbook()
ws = wb.create_sheet('Аркуш № 1')
ws = wb.create_sheet('Аркуш № 2')
ws = wb.active
ws.title = 'SAMPLE'
ws.cell(1,4).value = 'Value'
ws.cell(1,4).alignment = CENTER
ws.cell(2,4).fill = fillGrey
ws['C7'] = 'qwerty'
ws.freeze_panes='A2'
ws.row_dimensions.group(5, 8, hidden=True)
ws.column_dimensions['K'].value = 3
```
