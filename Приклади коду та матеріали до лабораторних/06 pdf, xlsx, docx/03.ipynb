# pypdf

[pypdf](https://pypdf.readthedocs.io/en/stable/) &mdash; бібліотека для роботи з pdf-файлами.

Сайт ILovePDF вміє багато, але є недоліки (наприклад, залишає службові закладки; проблематично встановлювати власні закладки (принаймні в безкоштовній версії)).

```python
from pathlib import Path
from pypdf import PdfReader, PdfWriter
import pypdf, sys
print('pypdf', pypdf.__version__)
```

```python
root = Path('.')
lectures = Path('..','..','..', 'C++1','lectures')
dest_lectures = root.joinpath('data','all_lectures.pdf')
sample = root.joinpath('data','69757_ОПП_F3_Б_Інформатика_2025.PDF')
sample_dest = sample.with_name(sample.stem+'_rotated.pdf')
```

```python
import re
topics = list()
for p in lectures.glob('*/lect*.pdf'):
    # print(p.parent.name)
    m = re.match(r'L(\d+)_(?:(\d)_)?(.+)', p.parent.name)
    topics.append((int(m[1]), m[2] if m[2]is not None else '', m[3], p))
    # print(topics[-1])
topics.sort()
sample2 = topics[7][-1]
# sample2
```

[PdfReader](https://pypdf.readthedocs.io/en/stable/modules/PdfReader.html) &mdash; клас, що забезпечує читання pdf-файлу.
<br>
[PdfWriter](https://pypdf.readthedocs.io/en/stable/modules/PdfWriter.html) &mdash; клас, що забезпечує формування та запис pdf-файлу.

<code>PdfReader.get_num_pages()</code>
<br>
<code>PdfReader.page[num]</code>

```python
reader = PdfReader(sample)
reader.get_num_pages()
```

<code>PdfWriter.append</code> &mdash; додає сторінки (одну, документ, кілька з документа) у кінець
<br>
<code>PdfWriter.merge</code> &mdash; додає сторінки (одну, документ, кілька з документа), можна задавати позицію
<br>
<code>PdfWriter.append_pages_from_reader</code>
<br>
<code>PdfWriter.write</code> &mdash; записує у файл


```python
writer = PdfWriter()
writer.add_metadata(
    {
        "/Author": "Tatiana",
        "/Producer": "LaTeX",
        "/Title": "All the lectures",
        "/Subject": "Programming",
        "/Keywords": "C++",
    }
)
prev = 0
for i, el in enumerate(topics):
    mark = f'{el[0]}' + (f'-{el[1]}' if el[1] else '') + ' ' + el[2]
    print(mark)
    reader = PdfReader(el[-1])
    writer.append_pages_from_reader(reader)
    # writer.append(el[-1]) # дуже схоже на попереднє, але різниця існує
    writer.add_outline_item(mark, prev)
    prev = len(writer.pages)
    if i>4: break
with open(dest_lectures, "wb") as fp:
    writer.write(fp)
```

```python
writer = PdfWriter()
writer.add_metadata(
    {
        "/Title": "ОПП Інформатика 2025 rotated",
    }
)
writer.append(sample)
to_rotate = list(range(2,9))+[22,23]+list(range(25,33))
for i in to_rotate:
    writer.pages[i].rotate(90)
writer.write(sample_dest);
```
