```python
from pathlib import Path
import pymupdf, sys
print('pymupdf', pymupdf.__version__)
```

```python
root = Path('.')
data = root.joinpath('data')
pdfsample = data.joinpath('104.pdf')
```

```python
doc = pymupdf.open(pdfsample) 
doc.page_count
```

```python
page = doc[1]
```

```python
page.get_text()
```

```python
print(page.get_textpage_ocr().extractText())  # tesseract, pip install mytesseract
```

```python
pix = page.get_pixmap()
pix.save(pdfsample.with_name(pdfsample.stem+ f"page-{page.number+1}.png"))
```
