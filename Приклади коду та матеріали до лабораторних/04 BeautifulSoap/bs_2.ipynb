# Розбір html-файлів - 2

```python
import pandas as pd
import html5lib
from bs4 import BeautifulSoup
from pathlib import Path
import lxml
```

```python
fname = Path('prepare', 'bbb.html')
with open(fname,  encoding='utf-8') as f:
    data = f.read()
```

```python
soup = BeautifulSoup(data, 'html5lib')
```

Наявну у html-файлі таблиці можна завантажити в таблицю вручну, якщо спочатку завантажити її у структуру даних Python, а потім за ціює структурою побудувати об'єкт типу <code>DataFrame</code>.

Тег <code>tbody</code> позначає тіло таблиці, тег <code>tr</code> &mdash; рядок таблиці, <code>th</code> &mdash; клітину заголовку, <code>td</code> &mdash; звичайну клітину.

```python
columns = ["Date", "Open", "High", "Low", "Close", "Volume", 'Adj_Close']
loaded = {el:[] for el in columns}
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    if (col != []):
        for i, el in enumerate(columns):
            loaded[el].append(col[i].text)

netflix_data = pd.DataFrame(loaded)
netflix_data.head()
```

Насправді, таблиці можна завантажувати напряму з <code>html</code>-файлу чи навіть з URL. За умовчанням використовується парсер <code>lxml</code>.
