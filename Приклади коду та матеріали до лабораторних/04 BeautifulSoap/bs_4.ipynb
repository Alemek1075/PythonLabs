## Finally

```python
def get_viewstate(soup):
    return soup.find('input', attrs={'id':'__VIEWSTATE'}).attrs['value']
```

```python
def get_authors(soup):
    form = soup.find_all('form')[0]
    authors = []
    for op in form.find(id='author').find_all('option'):
        if not op.has_attr('value'):continue
        authors.append(op.attrs['value'])
        # print(op.attrs['value'])
    default_tag = form.find(attrs={'name':'tag'}).option.text
    return authors, default_tag
```

```python
def get_author_tags(author, default_tag, viewstate):
    formData={'author':author, 'tag':default_tag, '__VIEWSTATE':viewstate}
    response = requests.post(url2, data=formData)
    soup = BSoup(response.text, 'html5lib') 
    form = soup.find_all('form')[0]
    selected = form.find(attrs={'name':'author'}).find('option', attrs={'selected':''}).attrs['value']
    if selected!=author:
        raise Exception('Incorrect author!')
    tags = []
    for op in form.find(id='tag').find_all('option'):
        if not op.has_attr('value'):continue
        tags.append(op.attrs['value'])
        # print(op.attrs['value'])
    return tags
```

```python
def get_quotes(author, tag, viewstate):
    formData={'author':author, 'tag':tag, '__VIEWSTATE':viewstate}
    response = requests.post(url2, data=formData)
    soup = BSoup(response.text, 'html5lib') 
    form = soup.find_all('form')[0]
    selected = form.find(attrs={'name':'author'}).find('option', attrs={'selected':''}).attrs['value']
    if selected!=author:
        raise Exception('Incorrect author!')
    selected = form.find(attrs={'name':'tag'}).find('option', attrs={'selected':''}).attrs['value']        
    if selected!=tag:
        raise Exception('Incorrect tag!')
    
    # response = requests.post(url2, data=formData)
    soup.find(class_='results').find_all(class_='quote')
    quotes = []
    for res in soup.find(class_='results').find_all(class_='quote'):
        quotes.append(res.find(class_='content').text)
    return quotes
```

```python
url0 = 'http://quotes.toscrape.com'
url1 = 'http://quotes.toscrape.com/search.aspx'
url2 = 'http://quotes.toscrape.com/filter.aspx'
response  = requests.get(url1)
soup = BSoup(response.text, 'html5lib') 
viewstate = get_viewstate(soup)
authors, default_tag = get_authors(soup)
res = dict()
for author in authors:
    tags = get_author_tags(author, default_tag, viewstate)
    res[author] = {tag:[] for tag in tags}
    for tag in tags:
         res[author][tag] = get_quotes(author, tag, viewstate)
```

```python
for a in res:
    print(a)
    for t in res[a]:
        print('---', t)
        for q in res[a][t]:
            print('--- --- ', q)
```
