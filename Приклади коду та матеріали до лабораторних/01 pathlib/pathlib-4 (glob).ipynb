## glob

```python
x = Path('..')
x.glob('*')
```

```python
for el in x.glob('*'):
    print(el)
```

```python
for el in x.glob('*.py'):
    print(el)
```

```python
for el in x.glob('*.*py*'):
    print(el)
```

```python
for el in x.glob('i*.*py*'):
    print(el)
```

```python
for el in x.glob('*.?????'):
    print(el)
```

```python
for el in x.glob('*[ys]'):  #[перелік символів] один з символів переліку
    print(el)
```

```python
for el in x.glob('[a-k]*.*'): # a-k усі символи від a до k включно
    print(el)
```

```python
for el in x.glob('[a-k]*[vm].*'):
    print(el)
```

```python
for el in x.glob('[a-k]*[vmy]'): #діапазони можна використовувати разом з переліком окремих символів
    print(el)
```

```python
for el in x.glob('[!a-k]*[vmy]'): # ! усі, крім
    print(el)
```

```python
for el in x.glob('*y*/'):  # 3.11   / на кінці шаблону: тільки папки
    print(el)
```

```python
for el in x.glob('**'):   
    print(el)
```

```python
for el in x.glob('**/'):
    print(el)
```

```python
for el in x.glob('**/*'):
    print(el)
```

```python
for el in x.glob('**/*.txt'):
    print(el)
```

```python
for el in x.glob('**/*.???'):
    print(el)
```

```python
for el in work.glob('**/*.*'):
    print(el)
```

```python
for el in work.glob('**/TestDir/*/*.txt'):
    print(el)
```
