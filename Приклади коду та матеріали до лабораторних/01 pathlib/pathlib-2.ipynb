## resolve, samefile

```python
p = Path('X/1/1.txt')  # relative path
print(p)
print(p.resolve())     # corresponding absolute path
```

```python
p = Path('./X/1/1.txt')  # relative path; with ./ is the same as previous but more safety
print(p)                 
print(p.resolve())       # corresponding absolute path
```

```python
show('.')
```

```python
p = Path('X/1/../2/3.txt')    # .. - parent (one level up)
print(p)                 
print(p.resolve())      
```

```python
p = Path('X/1/../2/./3.txt')  # . - current (this level), but not current work dir
print(p)                 
print(p.absolute())   
print(p.resolve())      
```

```python
p4 = Path('X/1/4.txt')  # 4.txt is a symlink to 1.txt
print(p4)                 
print(p4.absolute())   
print(p4.resolve())      
```
