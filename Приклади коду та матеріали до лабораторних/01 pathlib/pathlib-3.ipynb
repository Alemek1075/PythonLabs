## group operations

```python
tmp = Path(samples,'tmp')
Path(tmp,'1/22/333').mkdir(parents=True)
show('.')
```

```python
try:
    tmp.rmdir()
except BaseException as e:
    print(e)
```

```python
shutil.rmtree(tmp, ignore_errors=True)
show('.')
```

```python
shutil.copytree('X', 'Samples/X')
show('.')
```

## moving

```python
show(samples)
```

```python
p = samples.joinpath('X','1','1.txt')
new_p = samples.joinpath('X','5.txt')
p.replace(new_p)
show(samples)
```

```python
new_p.rename(p) # this isn`t the same as replace, there are nuances that depends on OS and filesystem
show(samples)
```

```python
shutil.rmtree(samples.joinpath('X'), ignore_errors=True)
show('.')
```
