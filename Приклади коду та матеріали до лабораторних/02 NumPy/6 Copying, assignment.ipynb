# Samples of copying and assignment

```python
import numpy as np
import sys
print(sys.platform, sys.version, f'numpy version {np.__version__}', sep='\n')
```

## Recollection lists copying

```python
a_list = [[1, 2], [3, 4]]
a_list
```

```python
b_list = a_list
a_list is b_list
```

```python
d_list = a_list.copy()
d_list[0][0] = 13
print(a_list)
```

## ndarray copying and assignment

```python
a = np.array([[1, 4], [7, 32]])
a
```

```python
b = a
c = a.copy()
print(b is a)
print(c)
```

```python
b = a.T  # побудували транспоновану матрицю
b
```

```python
c[0, 0] = -8
print('a:\n', a)
print('b:\n', b)
print('c:\n', c)
```

```python
a[1] = [23, 34]
print('a:\n', a)
print('b:\n', b)
print('c:\n', c)
```

```python
a[0] = 13   # broadcasting
print('a:\n', a)
```

```python
a.T[0] = 9
print('a:\n', a)
```
