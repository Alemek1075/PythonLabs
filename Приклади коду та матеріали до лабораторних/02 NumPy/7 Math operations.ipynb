## Linear algebra

```python
#!pip install scipy
```

```python
import scipy.linalg as la
```

```python
a = np.array([[1, 0, 0, 0], [1, 1, 0, 0], [1, 2, 1, 0], [1, 3, 3, 1]])
a
```

```python
a1 = la.inv(a)
a1
```

```python
a2 = np.array(a1, dtype=int)
a2
```

```python
a2@a == np.identity(4)
```

## Linear system ax=b solving

```python
a = np.identity(3, dtype=int)
b = np.arange(1, 4)
np.hstack((a, b.reshape(3, 1)))  # only for the sake of nice printing
```

```python
la.solve(a, b)
```
