## linear index calculating

Let build matrix of size $3\times 4$.

```python
a = np.array(range(12)).reshape(3, 4)
a
```

```python
a.flags
```

Now, we let transpose our matrix. The result will be of size $4\times 3$.

```python
b = a.T
b
```

```python
b.flags
```

```python
b.item(2)
```

```python
a.item(2)
```
