## Broadcasting

```python
v = np.arange(1, 4, dtype=int)
v_1_3 = v.reshape(1, 3) 
v_3_1 = v.reshape(3, 1) 
print('v =', v)
print('v_1_3 =', v_1_3)
print('v_3_1 =\n', v_3_1)
```

```python
ones3 = np.ones((3, 3), dtype=int)
print('ones3 =\n', ones3)
```

```python
ones3 * v_1_3
```

```python
ones3 * v_3_1
```

```python
ones3 * v
```

```python
v_1_3 + v_3_1
```

## Indexing by sequence 

```python
a = np.arange(16, dtype=int).reshape(4, -1)
a
```

```python
a[[1, 2], [0, 3]]
```

```python
a[(1, 2), (0, 3)]
```

```python
a[[0, 3]]
```

```python
tmp = a[[0, 3]][:, [0, 2]]
tmp
```

## Joining and splitting

```python
x = np.arange(3)
a = np.arange(1, 10).reshape(3, -1)
print('x =', x)
print('a =\n', a)
```

```python
np.concatenate((x, x))
```

```python
np.hstack((x, x))
```

```python
print(x[np.newaxis, :])
np.hstack((x[np.newaxis, :], x[np.newaxis, :]))
```

```python
np.concatenate((x[np.newaxis, :], x[np.newaxis, :]), 1)
```

```python
np.vstack((x, x))
```

```python
np.vstack((x[np.newaxis, :], x[np.newaxis, :]))
```

```python
np.concatenate((x[np.newaxis, :], x[np.newaxis, :]), 0)
```

```python
b = np.concatenate((np.tril(a), np.triu(a)), 1)
b
```

```python
tmp = np.hsplit(b, (3,))
print(*tmp, sep='\n\n')
```

```python
tmp = np.hsplit(b, (2, 5))
print(*tmp, sep='\n\n')
```

```python
tmp = np.hsplit(b, 3)
print(*tmp, sep='\n\n')
```

```python
tmp = np.vsplit(b, (1,))
print(*tmp, sep='\n\n')
```
