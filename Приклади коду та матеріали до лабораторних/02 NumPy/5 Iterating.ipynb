# Samples of iterating

```python
import numpy as np
import sys
print(sys.platform, sys.version, f'numpy version {np.__version__}', sep='\n')
```

```python
a = np.array([[1, 4], [7, 32]])
a
```

```python
for el in a: 
    print(el)
```

```python
for el in a.flat: 
    print(el)
```

```python
type(el)
```

```python
for i, el in enumerate(a): 
    print(i, el)
```

```python
for i, el in np.ndenumerate(a): 
    print(i, el)
```
