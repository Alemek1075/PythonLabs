# Samples of ndarray creating

```python
import numpy as np
import sys
print(sys.platform, sys.version, f'numpy version {np.__version__}', sep='\n')
```

## array

```python
a = np.array(range(5))
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([1, 3, 5, 2])
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([[1, 3], [5, 2]], dtype=int)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([[1, 3], [5, 2], [1, 1]])
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([[1, 3.5], [5, 2]])
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([[1, 3.5],[5, 2+1j]])
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([1,3,5,2], dtype=int)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([1,3,5,2], dtype=np.int64)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([1, 3, 5, 2], dtype=float)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([1, 3, 5, 2], dtype=np.float64)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([1, 3, 5, 2], dtype=np.double)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([(1, 3), [5, 2]], dtype=float)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.array([(1.2, 3), [5, 2]], dtype=int)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

## arange

```python
a = np.arange(10)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.arange(0.0, 1.0, 0.1)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.arange(0.0, 1.0, 0.999_999_999)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.arange(0.0, 1.001, 0.1)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

## linspace

```python
a = np.linspace(0.0, 1.0, 11)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.linspace(0, 9, 10)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

```python
a = np.linspace(0, 9, 10, endpoint=False)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```

## fromiter

```python
a = np.fromiter((i*i for i in range(5)), dtype=np.int64)
print(a.ndim, a.shape, a.dtype, sep='; ')
print(a)
```
