# Samples of reshaping

```python
import numpy as np
import sys
print(sys.platform, sys.version, f'numpy version {np.__version__}', sep='\n')
```

```python
a = np.arange(1, 13)
a
```

```python
b = a.reshape(2, 6)
b
```

```python
c = b.reshape(3, 4)
c
```

```python
a.T
```

```python
d = c.T
d
```

```python
e = d.reshape(2, 6)
e
```

```python
b.shape = (3, 4)
b
```

```python
try:
    d.shape = (2, 6)
except BaseException as e:
    print(e)
```

```python
b = b.reshape(2, -1)   # розмір останнього виміру буде обчислено автоматично
b
```

```python
b.shape = (3, -1)
b
```

```python
e = b.flatten()
print('e:', e)
```

```python
b.reshape(b.size)
```

```python
b.shape = (b.size,)
```

```python
b.shape = (3, 4)
b
```

```python
li = b.tolist()
print(li)
print(type(li[0][0]))
```
