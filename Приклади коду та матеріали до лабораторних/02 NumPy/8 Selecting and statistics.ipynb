## Samples of selecting

### Sample 1

```python
a = np.arange(1, 17, dtype=int).reshape(4, -1)
a
```

```python
a%3 == 0
```

```python
a[a%3 == 0]
```

```python
np.logical_and(a%3 == 0, a%5 == 1)
```

```python
(a%3 == 0) & (a%5 == 1)
```

```python
np.logical_or(a%3 == 0, a%5 == 1)
```

```python
(a%3 == 0) | (a%5 == 1)
```

```python
a[(a%3==0) & (a%5==1)]
```
