### Перетворення

<b>Переупорядкування стовпчиків, зміна індексу</b>

```python
data = data.reindex(columns=('id', 'v1', 'v2', 'v3'))  # copy = True by default
data.head(2)
```

```python
data.set_index('id', inplace=True)
data.head(2)
```

```python
data.reset_index(inplace=True)
data.head(2)
```

<b>Додавання та вилучення стовпчиків</b>

```python
data['v4'] = data.v1 + data.v2
data.head(3)
```

```python
# del data['v4']
data.drop(columns='v4', inplace=True)
data.head(3)
```

```python
data['x'] = 0
data['y'] = 0
data.head(2)
```

```python
data.drop(columns=['x', 'y'], inplace=True)
data.head(3)
```

<b>Сортування даних</b>

```python
data['total'] = data['v1']+data['v2']+data['v3']
data.head(5)
```

```python
data.sort_values(by=['total', 'id'], ascending=[False, True], inplace=True)
data
```

```python
data.index
```

```python
data.sort_index(inplace=True)
data
```

### Вибірки даних

```python
(data['v1'] >= 3)  # можна і data.v1, і data['v1']
```

```python
(data['v1'] >= 3) & (data['v3'] > 22)
```

```python
np.logical_and((data['v1'] >= 3), (data['v3'] > 22))
```

```python
data[(data['v1'] >= 3) & (data['v3'] > 22)]
```

```python
data[(data['v1']<3) | (data['v3']==20)]
```
