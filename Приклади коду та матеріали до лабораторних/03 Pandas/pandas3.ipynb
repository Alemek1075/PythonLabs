# pandas.DataFrame: Робота з рядковими значеннями

```python
import numpy as np
import pandas as pd
import sys
print(sys.platform, sys.version, f'numpy version {np.__version__}',  f'pandas version {pd.__version__}',sep='\n')
```

Складемо таблицю <code>data</code> з рядковими даними. Оскільки далі над нею будуть виконуватись різноманітні перетворення, то створимо її копію у змінній <code>orig</code>, щоб не повторювати процес створення.

```python
import random
n = 10
ids = [f'var {i}' for i in range(1, n+1)]
v1 = [random.randint(0, 9) for _ in range(n)]
v2 = [random.randint(10, 19) for _ in range(n)]
v3 = (random.randint(20, 29) for _ in range(n))
raw_data = {'id':ids, 'v1':v1, 'v2':v2, 'v3':v3}
data = pd.DataFrame(raw_data)
orig = data.copy()
data.head()
```

Стовпчик <code>id</code> таблиці містить рядкові дані, що позначають номери варіантів.

```python
data.info()
```

Відсортуємо таблицю за значеннями стовпчика  <code>id</code>. 

```python
data.sort_values('id', inplace=True)
data.head(3)
```

Оскільки значеннями є рядки, то сортування виконано в лексикографічному порядку. Це не зовсім влаштовує, адже <code>var 10</code> має розміщуватися після <code>var 9</code>, а не після <code>var 1</code>.

Отримане сортування обумовлено саме тим, що сортувалися рядкові дані.

Щоб отримати сортування за номерами варіантів, необхідно виконати перетворення. Наприклад, побудувати числовий стовпчик з номерами варіантів.

Проведемо попередні дослідження цього питання.

```python
data.id.str.findall(r'[var]')  # v|a|r DM: v+a+r  [a-z]  [^abc] [a-zA-Z]
```

```python
data.id.str.findall(r'(\s*(var)\s*(\d+)\s*)')   #\s  \d # [abc]+  [0-9]    a|ab
```

```python
data.id.str.findall(r'\s*var\s*(\d+)\s*').str[0]
```

```python
data.id.str.match(r'\s*var\s*\d+\s*')
```

```python
#import re
#r1 = re.compile(r'\s*var\s*(\d+)\s*')
#def f(s, r1):
#    res = re.fullmatch(r1, s)
#    return res[1]

#data.id.apply(f, args=(r1,))
```

```python
data.id = data.id.str.findall(r'\s*var\s*(\d+)\s*').str[0]
data.head()
```

```python
data.sort_values(by='id', inplace=True)
data.head()
```

```python
data.info()
```

```python
data.id = data.id.astype('int64')
data.sort_values(by='id', inplace=True)
data.head()
```
