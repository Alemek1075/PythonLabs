```python
import csv

def load_data(fname, skip_headers=True):
    data = list()
    with open(fname, newline='', encoding='utf-8') as f:
        r = csv.reader(f)
        if skip_headers:
            next(r)
        for row in r:
            if row:
                data.append(row)
    return data
```

Файл <code>population.csv</code> містить чисельність населення деяких міст України станом на 2014 рік (за даними Wikipedia).

```python
tmp = load_data(Path('prepare', 'population.csv'))
tmp[:5]
```

```python
raw_data = {el[0].strip():int(el[1]) for el in tmp[:5]}
raw_data
```

```python
population = pd.Series(raw_data)
population
```

```python
population = pd.Series(data=(int(el[1]) for el in tmp), index=(el[0].strip() for el in tmp))
population.head()
```

```python
population.tail()
```

```python
population.head(3)
```

```python
population.array
```

```python
population.index
```

```python
population.name = 'population'
population.index.name = 'city'
population.head()
```

### Індексація та зрізи

<code>at</code>,<code>loc</code> використовують індекси (мітки даних)<br>
<code>iat</code>,<code>iloc</code> використовують номери

```python
population.at['Київ'], population.loc['Київ']
```

```python
population.iat[0], population.iloc[0]
```

<code>...at</code> надають доступ до одного елемента даних<br>
<code>...loc</code> надають доступ до певної частини даних (але може складатись з одного елемента)

Зрізи можуть бути по індексах та по номерах.<br>
Зрізи по номерах не включають праву межу, [from, to).<br>
Зрізи по індексах <b>включають</b> праву межу, [from, to].<br>

```python
population.iloc[0:3]
```

```python
population.loc['Київ':'Дніпро']
```

```python
population.loc[['Київ', 'Дніпро']]
```

```python
population.iloc[[1, 2, 7]]
```

<b>Індексація</b> [] можлива по індексу (по номерах планують прибрати), зрізи можливі по індексах та по номерах.

```python
population['Київ']
```

```python
population[0:3]
```

```python
population['Київ':'Дніпро']
```

### Оброблення відсутніх даних

Файл <code>area.csv</code> містить площу деяких міст України (за даними з інтернету). Але дані заповнені не для всіх міст, що наявні в таблиці. Замість відсутніх значень під час завантаження підставляється <code>np.nan</code> (not a number, виділене значення дійсного числового типу з плаваючою крапкою).

```python
tmp = load_data(Path('prepare', 'area.csv'))
area = pd.Series((float(el[1]) if el[1] and not el[1].isspace() else np.nan for el in tmp), 
                 index=(el[0].strip() for el in tmp))
area.head(5)
```

Використаємо розширений тип.

```python
tmp = load_data(Path('prepare', 'area.csv'))
area = pd.Series((float(el[1]) if el[1] and not el[1].isspace() else np.nan for el in tmp), 
                 index=(el[0].strip() for el in tmp), dtype='Float64')
area.head(5)
```

Якщо відсутнє значення дійсного типу може бути зображене дійсним "нечислом", то у випадку цілих типів, наявність спеціального значення для позначення відсутності значення є важливою.

```python
tmp = load_data(Path('prepare', 'area.csv'))
area_int = pd.Series((int(float(el[1])) if el[1] and not el[1].isspace() else None for el in tmp), 
                 index=(el[0].strip() for el in tmp), dtype='Int64')
area_int.head(5)
```

Перевіримо на відсутність значень. 

```python
area.isna()
```

Відсутні значення можна вилучити з послідовності (dropna) або можна заповнити (fillna).

```python
area.dropna(inplace=True)
area
```

### Векторні операції над послідовностями з різними індексами
