## Read csv

```python
def read_csv(fname, newline='', encoding='utf-8'):
    with open(fname, newline=newline, encoding=encoding) as f:  
        r = csv.reader(f)
#         headers = next(r)
        for el in r:
            print(el)
```

```python
def read_csv_strict(fname, newline=''):
    with open(fname, newline=newline) as f:
        r = csv.reader(f, strict=True)
        for el in r:
            print(el)
```

## Headers processing

```python
def read_csv_with_headers(fname, newline=''):
    with open(fname, newline=newline) as f:  
        r = csv.reader(f)
        headers = next(r)
        for el in r:
            print(el)
```

```python
headers = ['x', 'f(x)']
data = [{headers[1]:x*x, headers[0]:x} for x in {1, 2, 5}]
data
```

```python
fname = iosamples.CSV_HEADERS
with open(fname,'w',newline='') as f:
    w = csv.DictWriter(f, headers)
    w.writeheader()
    for el in data:
        w.writerow(el)
read_csv(fname)
```
